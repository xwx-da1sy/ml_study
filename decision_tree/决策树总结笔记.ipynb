# 决策树总结笔记

这份笔记总结目前学过的决策树内容：基本概念、信息熵、条件熵、信息增益、ID3、C4.5、CART，以及 C5.0 的位置。

## 1. 决策树解决什么问题

决策树主要用于**有监督学习**。

- 分类问题：预测类别，例如是否批准贷款、是否打球、是否患病。
- 回归问题：预测连续值，例如房价、收入、销量。

训练数据必须包含特征和标签：

$$X \rightarrow y$$

例如：

| 有工作 | 有房产 | 是否批准贷款 |
| --- | --- | --- |
| 是 | 是 | 批准 |
| 否 | 否 | 拒绝 |

决策树的本质是：不断选择一个特征来提问，把数据分得越来越纯。

## 2. 信息熵

信息熵衡量一个数据集的混乱程度。

$$H(D)=-\sum_{k=1}^{K}p_k\log_2 p_k$$

其中 $p_k$ 是第 $k$ 类样本所占比例。

直觉：

- 全是一类：$H(D)=0$，最纯。
- 类别越平均：$H(D)$ 越大，越混乱。

例如贷款数据中：批准 5 条，拒绝 3 条，共 8 条。

$$P(批准)=\frac{5}{8},\quad P(拒绝)=\frac{3}{8}$$

$$H(D)=-\frac{5}{8}\log_2\frac{5}{8}-\frac{3}{8}\log_2\frac{3}{8}$$

$$H(D)\approx0.9544$$

## 3. 条件熵

条件熵表示：按照某个特征分组以后，标签还剩多少不确定性。

$$H(D|A)=\sum_{v\in Values(A)}\frac{|D_v|}{|D|}H(D_v)$$

计算步骤：

1. 按特征 $A$ 分组。
2. 分别计算每个分组内部的熵。
3. 按每个分组的样本占比加权平均。

例如按 `有工作` 分组：

| 分组 | 样本数 | 批准 | 拒绝 | 组内熵 |
| --- | ---: | ---: | ---: | ---: |
| 有工作=是 | 4 | 4 | 0 | 0 |
| 有工作=否 | 4 | 1 | 3 | 0.8113 |

$$H(D|有工作)=\frac{4}{8}\times0+\frac{4}{8}\times0.8113$$

$$H(D|有工作)=0.4056$$

注意：纯节点的熵为 0，可以省略，但第一次写最好标出来。

## 4. 信息增益

信息增益表示：使用某个特征划分以后，不确定性减少了多少。

$$Gain(D,A)=H(D)-H(D|A)$$

例如：

$$Gain(D,有工作)=0.9544-0.4056$$

$$Gain(D,有工作)=0.5488$$

信息增益越大，说明这个特征让数据变得越纯。

## 5. ID3 决策树

ID3 的核心规则：

$$\text{选择信息增益最大的特征进行划分}$$

例如：

$$Gain(D,有工作)=0.5488$$

$$Gain(D,有房产)=0.3476$$

因为：

$$0.5488>0.3476$$

所以 ID3 会优先选择 `有工作`。

优点：

- 思路清楚，适合理解决策树。
- 每一步都有明确指标。
- 生成的树比较容易解释。

缺点：

- 偏爱取值很多的特征。
- 主要适合离散特征。
- 容易过拟合。

## 6. C4.5 决策树

C4.5 是 ID3 的改进版。ID3 使用信息增益，C4.5 使用信息增益率。

$$GainRatio(D,A)=\frac{Gain(D,A)}{SplitInfo(D,A)}$$

其中：

- $Gain(D,A)$：这个特征让标签变纯了多少。
- $SplitInfo(D,A)$：这个特征把数据切得有多碎。

分裂信息：

$$SplitInfo(D,A)=-\sum_{v\in Values(A)}\frac{|D_v|}{|D|}\log_2\frac{|D_v|}{|D|}$$

例子：

| 特征 | 信息增益 $Gain$ | 分裂信息 $SplitInfo$ | 信息增益率 $GainRatio$ |
| --- | ---: | ---: | ---: |
| 有工作 | 0.5488 | 1.0000 | 0.5488 |
| 用户编号 | 0.9544 | 3.0000 | 0.3181 |

如果只看信息增益，`用户编号` 很容易被误选，因为每个用户编号都能把样本切得很纯。

但 C4.5 会用 $SplitInfo$ 惩罚这种“切太碎”的特征。

$$GainRatio(D,用户编号)=\frac{0.9544}{3}=0.3181$$

记忆：

- 信息增益：问“纯了多少”。
- 信息增益率：还要问“是不是切太碎了”。

## 7. CART 决策树

CART 全称是 Classification And Regression Tree。它既可以做分类，也可以做回归。

CART 的核心特点：

$$\text{每个节点都是二叉划分}$$

分类 CART 常用 Gini 衡量节点不纯度。

$$Gini(D)=1-\sum_{k=1}^{K}p_k^2$$

直觉：

- $Gini(D)=0$：节点最纯。
- $Gini(D)$ 越大：节点越混。

划分后的加权 Gini：

$$GiniIndex(D,A)=\frac{|D_1|}{|D|}Gini(D_1)+\frac{|D_2|}{|D|}Gini(D_2)$$

CART 选择：

$$\text{加权 Gini 最小的划分}$$

例如按 `有工作` 划分：

| 分支 | 样本数 | 批准 | 拒绝 | Gini |
| --- | ---: | ---: | ---: | ---: |
| 有工作=是 | 4 | 4 | 0 | 0 |
| 有工作=否 | 4 | 1 | 3 | 0.375 |

$$GiniIndex(D,有工作)=\frac{4}{8}\times0+\frac{4}{8}\times0.375$$

$$GiniIndex(D,有工作)=0.1875$$

按 `有房产` 划分：

$$GiniIndex(D,有房产)=0.3000$$

因为：

$$0.1875<0.3000$$

所以 CART 选择 `有工作`。

注意：CART 不是比较某一个分支的 Gini，而是比较整次划分后的加权 Gini。

## 8. 多叉树和 CART 的区别

ID3、C4.5 常见形式可以是多叉树。

```text
天气?
├── 晴
├── 阴
└── 雨
```

CART 固定是二叉树。

```text
天气 是否属于 {晴} ?
├── 是
└── 否
```

对比：

| 类型 | 一次划分 | 优点 | 缺点 |
| --- | --- | --- | --- |
| 多叉树 | 可分多个分支 | 直观，树可能更浅 | 分支容易爆炸，容易过拟合 |
| CART | 只能分两个分支 | 结构统一，适合连续特征和剪枝 | 树可能更深 |

## 9. 连续特征怎么处理

CART 很适合处理连续特征。

例如年龄：

$$22,\ 25,\ 28,\ 35,\ 40$$

候选阈值通常取相邻值的中点：

$$\frac{22+25}{2}=23.5$$

$$\frac{25+28}{2}=26.5$$

$$\frac{28+35}{2}=31.5$$

$$\frac{35+40}{2}=37.5$$

每个阈值形成一个二叉划分：

$$年龄\le 26.5$$

然后计算每个阈值的加权 Gini，选择最小的。

## 10. C5.0 决策树

C5.0 是 C4.5 的进一步改进版。

$$C5.0\approx\text{更快、更省内存、更工程化的 C4.5 增强版}$$

常见特点：

- 使用类似 C4.5 的信息增益率思想。
- 支持生成决策树或规则集。
- 支持 boosting。
- 支持样本权重。
- 支持误分类代价。
- 支持特征筛选。
- 通常比 C4.5 速度更快，树更小。

注意：C5.0 的大体思想公开，但很多工程细节没有像 C4.5 那样完整公开成经典教材公式。

## 11. 总体对比

| 算法 | 选择划分的指标 | 树结构 | 主要特点 |
| --- | --- | --- | --- |
| ID3 | 信息增益 | 多叉较常见 | 简单直观，但偏爱多取值特征 |
| C4.5 | 信息增益率 | 多叉较常见 | 缓解多取值特征偏好 |
| CART | Gini 或平方误差 | 二叉树 | 工程常用，适合分类和回归 |
| C5.0 | 信息增益率等改进 | 树或规则集 | C4.5 的增强版，支持 boosting 等 |

## 12. 最容易混的点

### 条件熵和划分后 Gini 很像

它们都要先算每个子节点的不纯度，再按样本数加权平均。

$$H(D|A)=\sum_v\frac{|D_v|}{|D|}H(D_v)$$

$$GiniIndex(D,A)=\sum_v\frac{|D_v|}{|D|}Gini(D_v)$$

区别是：

- ID3/C4.5 用熵。
- CART 分类树用 Gini。

### 信息增益越大越好，GiniIndex 越小越好

$$Gain(D,A)=H(D)-H(D|A)$$

信息增益表示减少的不确定性，所以越大越好。

$$GiniIndex(D,A)$$

GiniIndex 表示划分后的不纯度，所以越小越好。

### CART 是二叉树

即使一个特征有多个取值，CART 也不会直接分成多叉，而是找一个二叉切法。

### $0\log 0$ 怎么处理

在信息熵里约定：

$$0\log_2 0=0$$

所以纯节点的熵为 0。

## 13. 一句话复习

$$\text{决策树的目标：把数据越分越纯。}$$

- ID3：用信息增益，选让熵下降最多的特征。
- C4.5：用信息增益率，避免被多取值特征骗。
- CART：用 Gini，选划分后最纯的二叉切法。
- C5.0：C4.5 的增强版，更快、更工程化。
