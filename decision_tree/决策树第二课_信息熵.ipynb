# 决策树第二课：信息熵（Entropy）

本课目标：理解信息熵如何衡量一个数据集的"混乱程度"，并为后续学习信息增益做好准备。

## 1. 为什么需要信息熵

决策树每一次分裂都希望让子节点中的样本类别尽可能一致，也就是让数据更"纯"。

- 如果一个节点中的样本全属于同一类，结果没有不确定性，节点最纯。
- 如果各类别数量接近，预测类别的不确定性较大，节点更混乱。

**信息熵**就是量化这种不确定性（或混乱程度）的指标。熵越小，节点越纯。

## 2. 信息熵公式

对于一个有 $k$ 个类别的数据集：

$$H(D) = -\sum_{i=1}^{k} p_i \log_2(p_i)$$

其中：

- $D$：当前数据集；
- $p_i$：第 $i$ 个类别在数据集中的比例；
- $H(D)$：数据集 $D$ 的信息熵。

约定：当 $p_i = 0$ 时，$p_i\log_2(p_i)$ 按 0 处理。

## 3. 两个直觉例子

### 例 1：节点完全纯

假设 10 个样本全为“是”：$p_{是}=1$。

$$H(D) = -(1\times\log_2 1) = 0$$

熵为 0，说明没有不确定性。

### 例 2：两类数量相同

假设 10 个样本中有 5 个“是”、5 个“否”：$p_{是}=p_{否}=0.5$。

$$H(D) = -(0.5\times\log_2 0.5 + 0.5\times\log_2 0.5) = 1$$

对于二分类问题，类别各占一半时熵最大，数据最混乱。

```python
import math

def entropy(labels):
    """计算标签列表的信息熵。"""
    total = len(labels)
    counts = {}

    for label in labels:
        counts[label] = counts.get(label, 0) + 1

    result = -sum(
        (count / total) * math.log2(count / total)
        for count in counts.values()
    )
    return 0.0 if result == 0 else result

pure_labels = ['是'] * 10
balanced_labels = ['是'] * 5 + ['否'] * 5

print(f'纯节点的信息熵：{entropy(pure_labels):.4f}')
print(f'均衡二分类节点的信息熵：{entropy(balanced_labels):.4f}')
```

预期输出：纯节点为 `0.0000`；两类各占一半时为 `1.0000`。

```python
# 观察二分类中类别比例变化对熵的影响
for positive_count in range(0, 11):
    labels = ['是'] * positive_count + ['否'] * (10 - positive_count)
    print(f'是：{positive_count:2d}，否：{10 - positive_count:2d}，熵：{entropy(labels):.4f}')
```

## 4. 熵与决策树分裂

设分裂前数据集的熵为 $H(D)$。使用某个特征分裂后，得到多个子集 $D_v$。

分裂后的加权平均熵为：

$$H_{after}(D) = \sum_v \frac{|D_v|}{|D|}H(D_v)$$

好的分裂会使分裂后的加权平均熵明显下降。熵减少的数量就是下一课要学习的**信息增益**：

$$InformationGain(D, A) = H(D) - H_{after}(D)$$

## 5. 本课小结

- 信息熵衡量数据集类别的不确定性。
- 熵越小，类别越集中，节点越纯。
- 对二分类问题，类别比例越接近 50% : 50%，熵越大。
- 决策树会偏好能让熵下降更多的特征；这种下降量称为信息增益。
