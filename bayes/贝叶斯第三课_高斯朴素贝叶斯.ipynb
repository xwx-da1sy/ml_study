# 贝叶斯第三课：高斯朴素贝叶斯

上一课的朴素贝叶斯例子用的是词频。

词频是离散的：

$$
中奖出现了 4 次
$$

$$
优惠出现了 6 次
$$

这种适合多项式朴素贝叶斯。

但是很多机器学习任务里的特征是连续值。

比如：

- 身高
- 体重
- 花瓣长度
- 血糖值
- 考试分数

这时候不能再简单数：

某个特征值出现了几次。

因为连续值几乎不会刚好重复。

比如身高：

$$
170.1,\quad 170.2,\quad 170.3
$$

它们都不完全一样。

所以高斯朴素贝叶斯要解决的问题是：

连续特征下，怎么计算：

$$
P(x_j\mid y)
$$

## 1. 朴素贝叶斯公式回顾

朴素贝叶斯分类规则是：

$$
\hat{y}
=
\arg\max_y
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
$$

这里最关键的是：

$$
P(x_j\mid y)
$$

也就是：

在类别 \(y\) 已经确定的情况下，第 \(j\) 个特征取到当前值的可能性。

上一课里，如果 \(x_j\) 是一个词，我们可以用词频估计：

$$
P(w\mid y)
=
\frac{N_{yw}+\alpha}{N_y+\alpha K}
$$

但是如果 \(x_j\) 是连续值，比如花瓣长度：

$$
x_j=5.1
$$

就不太适合问：

训练集中 \(5.1\) 出现了几次？

因为连续特征有无数种可能值。

所以我们改问：

如果类别是 \(y\)，那么这个连续特征大概服从什么分布？

高斯朴素贝叶斯的回答是：

假设它服从高斯分布，也就是正态分布。

## 2. 高斯分布在这里是什么意思？

高斯分布的密度函数是：

$$
p(x)
=
\frac{1}{\sqrt{2\pi\sigma^2}}
\exp\left(
-
\frac{(x-\mu)^2}{2\sigma^2}
\right)
$$

这里：

- \(\mu\)：均值，表示数据大概围绕哪个中心。
- \(\sigma^2\)：方差，表示数据分散得有多开。
- \(x\)：当前要计算可能性的特征值。

注意：

连续特征里严格来说不说：

$$
P(x_j=5.1\mid y)
$$

因为连续变量刚好等于某个点的概率通常是 \(0\)。

更准确地说，我们用的是概率密度：

$$
p(x_j\mid y)
$$

机器学习分类时，经常还是口头说：

$$
P(x_j\mid y)
$$

但你心里要知道：

连续特征这里用的是密度值，不是离散概率。

人话版：

高斯朴素贝叶斯不是数某个值出现了几次。

它先看某个类别下这个特征的均值和方差。

然后问：

当前这个值离该类别的中心近不近？

离中心越近，密度越大。

离中心越远，密度越小。

## 3. 参数 \(\mu\) 和 \(\sigma^2\) 怎么来？

假设类别是：

$$
y=c
$$

第 \(j\) 个特征在这个类别下有这些取值：

$$
x_{1j},x_{2j},\ldots,x_{m_cj}
$$

其中 \(m_c\) 表示类别 \(c\) 下有多少个样本。

均值估计为：

$$
\mu_{cj}
=
\frac{1}{m_c}
\sum_{i:y_i=c}
x_{ij}
$$

方差估计为：

$$
\sigma_{cj}^2
=
\frac{1}{m_c}
\sum_{i:y_i=c}
\left(
x_{ij}-\mu_{cj}
\right)^2
$$

说人话：

对每一个类别、每一个特征，都单独算一组均值和方差。

比如有两个类别：

$$
苹果,\ 香蕉
$$

有两个连续特征：

$$
重量,\ 甜度
$$

那么要分别算：

- 苹果的重量均值和方差。
- 苹果的甜度均值和方差。
- 香蕉的重量均值和方差。
- 香蕉的甜度均值和方差。

之后遇到一个新样本，就看它的每个特征，在不同类别对应的高斯分布下密度有多大。

## 4. 高斯朴素贝叶斯公式

对于一个样本：

$$
x=(x_1,x_2,\ldots,x_n)
$$

朴素贝叶斯分类规则是：

$$
\hat{y}
=
\arg\max_c
P(y=c)
\prod_{j=1}^{n}
P(x_j\mid y=c)
$$

高斯朴素贝叶斯把每个：

$$
P(x_j\mid y=c)
$$

换成高斯密度：

$$
p(x_j\mid y=c)
=
\frac{1}{\sqrt{2\pi\sigma_{cj}^2}}
\exp\left(
-
\frac{(x_j-\mu_{cj})^2}{2\sigma_{cj}^2}
\right)
$$

所以分类规则变成：

$$
\hat{y}
=
\arg\max_c
P(y=c)
\prod_{j=1}^{n}
\frac{1}{\sqrt{2\pi\sigma_{cj}^2}}
\exp\left(
-
\frac{(x_j-\mu_{cj})^2}{2\sigma_{cj}^2}
\right)
$$

这个式子看起来长，但意思很简单：

先看类别本身常不常见。

再看当前样本的每个连续特征，在这个类别的分布下像不像。

## 5. 为什么仍然叫朴素？

因为它还是用了条件独立假设：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
\prod_{j=1}^{n}
P(x_j\mid y)
$$

只是现在每个：

$$
P(x_j\mid y)
$$

不是用频数估计，而是用高斯密度估计。

也就是说：

高斯朴素贝叶斯的“高斯”指的是：

连续特征在每个类别下按正态分布处理。

高斯朴素贝叶斯的“朴素”指的是：

在类别已知时，各个特征之间仍然假设相互独立。

人话版：

它既没有复杂建模特征之间的关系，也没有学一条复杂边界。

它只是为每个类别、每个特征画一条正态曲线。

新样本来了以后，看它更像哪一类的这些正态曲线。

## 6. 手算例子：根据身高体重判断男女

我们做一个小例子。

特征有两个：

$$
x_1=身高
$$

$$
x_2=体重
$$

类别有两个：

$$
y=男
$$

$$
y=女
$$

训练数据如下：

| 编号 | 身高 | 体重 | 类别 |
|---:|---:|---:|---|
| \(1\) | \(170\) | \(65\) | 男 |
| \(2\) | \(180\) | \(75\) | 男 |
| \(3\) | \(175\) | \(70\) | 男 |
| \(4\) | \(160\) | \(50\) | 女 |
| \(5\) | \(165\) | \(55\) | 女 |
| \(6\) | \(155\) | \(48\) | 女 |

现在有一个新样本：

$$
x=(172,\ 68)
$$

问它更像男还是女。

先验概率：

$$
P(男)
=
\frac{3}{6}
=
0.5
$$

$$
P(女)
=
\frac{3}{6}
=
0.5
$$

## 7. 计算男类别的均值和方差

男类别的身高是：

$$
170,\quad 180,\quad 175
$$

身高均值：

$$
\mu_{男,身高}
=
\frac{170+180+175}{3}
=
175
$$

身高方差：

$$
\sigma_{男,身高}^2
=
\frac{(170-175)^2+(180-175)^2+(175-175)^2}{3}
$$

$$
\sigma_{男,身高}^2
=
\frac{25+25+0}{3}
=
16.6667
$$

男类别的体重是：

$$
65,\quad 75,\quad 70
$$

体重均值：

$$
\mu_{男,体重}
=
\frac{65+75+70}{3}
=
70
$$

体重方差：

$$
\sigma_{男,体重}^2
=
\frac{(65-70)^2+(75-70)^2+(70-70)^2}{3}
$$

$$
\sigma_{男,体重}^2
=
\frac{25+25+0}{3}
=
16.6667
$$

## 8. 计算女类别的均值和方差

女类别的身高是：

$$
160,\quad 165,\quad 155
$$

身高均值：

$$
\mu_{女,身高}
=
\frac{160+165+155}{3}
=
160
$$

身高方差：

$$
\sigma_{女,身高}^2
=
\frac{(160-160)^2+(165-160)^2+(155-160)^2}{3}
$$

$$
\sigma_{女,身高}^2
=
\frac{0+25+25}{3}
=
16.6667
$$

女类别的体重是：

$$
50,\quad 55,\quad 48

\sigma_{女,体重}^2

$$

## 9. 计算新样本在每个类别下的密度

新样本是：

$$
x=(172,\ 68)
$$

对男类别：

$$
p(172\mid 男)
=
\frac{1}{\sqrt{2\pi\times 16.6667}}
\exp\left(
-
\frac{(172-175)^2}{2\times 16.6667}
\right)
$$

$$
p(68\mid 男)
=
\frac{1}{\sqrt{2\pi\times 16.6667}}
\exp\left(
-
\frac{(68-70)^2}{2\times 16.6667}
\right)
$$

对女类别：

$$
p(172\mid 女)
=
\frac{1}{\sqrt{2\pi\times 16.6667}}
\exp\left(
-
\frac{(172-160)^2}{2\times 16.6667}
\right)
$$

$$
p(68\mid 女)
=
\frac{1}{\sqrt{2\pi\times 8.6667}}
\exp\left(
-
\frac{(68-51)^2}{2\times 8.6667}
\right)
$$

注意这个地方不要把密度当成普通概率。

密度值可以用来比较类别，但它本身不是“刚好等于某个点的概率”。

## 10. 计算类别分数

高斯朴素贝叶斯的分数是：

$$
Score(c)
=
P(y=c)
\prod_{j=1}^{n}
p(x_j\mid y=c)
$$

这里只有两个特征，所以：

$$
Score(男)
=
P(男)
p(172\mid 男)
p(68\mid 男)
$$

$$
Score(女)
=
P(女)
p(172\mid 女)
p(68\mid 女)
$$

因为新样本的身高 \(172\) 和体重 \(68\) 都更靠近男类别的中心：

$$
(175,\ 70)
$$

而离女类别中心：

$$
(160,\ 51)
$$

比较远。

所以男类别的分数会更大。

最终预测：

$$
\hat{y}=男
$$

## 11. 为什么用对数分数更好？

和上一课一样，高斯朴素贝叶斯也会把很多密度值相乘：

$$
P(y=c)
\prod_{j=1}^{n}
p(x_j\mid y=c)
$$

特征多时，数值可能非常小。

所以实际计算常用对数：

$$
\log Score(c)
=
\log P(y=c)
+
\sum_{j=1}^{n}
\log p(x_j\mid y=c)
$$

高斯密度取对数后：

$$
\log p(x_j\mid y=c)
=
-
\frac{1}{2}\log(2\pi\sigma_{cj}^2)
-
\frac{(x_j-\mu_{cj})^2}{2\sigma_{cj}^2}
$$

这个式子很重要。

它告诉我们两件事。

第一：

如果 \(x_j\) 离均值 \(\mu_{cj}\) 越远，那么：

$$
(x_j-\mu_{cj})^2
$$

越大，分数越低。

第二：

如果方差 \(\sigma_{cj}^2\) 很大，说明这个类别下该特征本来就很分散。

那么同样的偏离就没有那么严重。

人话版：

离类别中心越近，越像这个类别。

但如果这个类别本身就很分散，那么离中心远一点也还能接受。

## 12. 高斯朴素贝叶斯和前两种的区别

三种朴素贝叶斯可以这样对比。

| 类型 | 特征形式 | \(P(x_j\mid y)\) 怎么估计 |
|---|---|---|
| 多项式朴素贝叶斯 | 词频、计数 | 用次数比例估计 |
| 伯努利朴素贝叶斯 | 是否出现 | 用出现和不出现的比例估计 |
| 高斯朴素贝叶斯 | 连续数值 | 用高斯密度估计 |

它们共同使用：

$$
\hat{y}
=
\arg\max_y
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
$$

区别只在于：

$$
P(x_j\mid y)
$$

怎么得到。

所以不要把朴素贝叶斯理解成一个固定公式。

它更像是一套框架：

类别先验 \(P(y)\) 加上每个特征在该类别下的条件概率。

## 13. 本课小结

高斯朴素贝叶斯解决的是连续特征问题。

它的做法是：

第一，对每个类别、每个特征，分别计算均值：

$$
\mu_{cj}
$$

第二，对每个类别、每个特征，分别计算方差：

$$
\sigma_{cj}^2
$$

第三，用高斯密度估计当前特征值在该类别下的可能性：

$$
p(x_j\mid y=c)
=
\frac{1}{\sqrt{2\pi\sigma_{cj}^2}}
\exp\left(
-
\frac{(x_j-\mu_{cj})^2}{2\sigma_{cj}^2}
\right)
$$

第四，使用朴素贝叶斯分类规则：

$$
\hat{y}
=
\arg\max_c
P(y=c)
\prod_{j=1}^{n}
p(x_j\mid y=c)
$$

一句话：

高斯朴素贝叶斯就是把连续特征放进正态分布里，然后继续使用朴素贝叶斯的分类框架。

```python
import math
from collections import defaultdict

data = [
    (170, 65, "男"),
    (180, 75, "男"),
    (175, 70, "男"),
    (160, 50, "女"),
    (165, 55, "女"),
    (155, 48, "女"),
]

new_sample = (172, 68)

groups = defaultdict(list)
for height, weight, label in data:
    groups[label].append((height, weight))


def mean(values):
    return sum(values) / len(values)


def variance(values):
    mu = mean(values)
    return sum((x - mu) ** 2 for x in values) / len(values)


def gaussian_density(x, mu, var):
    return 1 / math.sqrt(2 * math.pi * var) * math.exp(-((x - mu) ** 2) / (2 * var))


total = len(data)
scores = {}

for label, samples in groups.items():
    prior = len(samples) / total
    heights = [s[0] for s in samples]
    weights = [s[1] for s in samples]

    height_mu = mean(heights)
    height_var = variance(heights)
    weight_mu = mean(weights)
    weight_var = variance(weights)

    height_density = gaussian_density(new_sample[0], height_mu, height_var)
    weight_density = gaussian_density(new_sample[1], weight_mu, weight_var)
    score = prior * height_density * weight_density
    log_score = math.log(prior) + math.log(height_density) + math.log(weight_density)
    scores[label] = score

    print(label)
    print("  身高均值:", height_mu)
    print("  身高方差:", height_var)
    print("  体重均值:", weight_mu)
    print("  体重方差:", weight_var)
    print("  身高密度:", height_density)
    print("  体重密度:", weight_density)
    print("  原始分数:", score)
    print("  对数分数:", log_score)

print("预测类别:", max(scores, key=scores.get))
```
