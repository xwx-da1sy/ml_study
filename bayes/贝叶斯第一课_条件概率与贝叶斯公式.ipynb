# 贝叶斯第一课：条件概率与贝叶斯公式

这一章开始换一种思路看机器学习。

前面的决策树、随机森林、AdaBoost、GBDT、XGBoost，核心问题更像是：

模型怎么切分？怎么投票？怎么一轮一轮修正错误？

贝叶斯要问的是另一个问题：

看到一些信息以后，某个结论成立的概率是多少？

比如：

看到一封邮件里有“优惠”“中奖”“链接”，它是垃圾邮件的概率是多少？

看到一个水果是红色，它是苹果的概率是多少？

这类问题的核心就是：

$$
P(类别 \mid 特征)
$$

也就是：

在已经看到特征的条件下，样本属于某个类别的概率。

## 1. 先把条件概率说清楚

条件概率的符号是：

$$
P(A\mid B)
$$

读作：

在 $B$ 已经发生的条件下，$A$ 发生的概率。

人话版：

原来我们在全部样本里看 $A$ 的比例。

现在我们先把范围缩小到 $B$ 发生的那些样本里，再看其中有多少也属于 $A$。

所以条件概率不是凭空来的。

它是在“缩小观察范围”。

如果 $B$ 发生了，那么我们只关心 $B$ 这个范围。

在 $B$ 里面，同时属于 $A$ 的部分是：

$$
A\cap B
$$

所以：

$$
P(A\mid B)
=
\frac{P(A\cap B)}{P(B)}
$$

这个公式的意思是：

在 $B$ 这块区域里，$A\cap B$ 占多大比例。

## 2. 条件概率公式怎么变成乘法公式？

从条件概率定义开始：

$$
P(A\mid B)
=
\frac{P(A\cap B)}{P(B)}
$$

两边同时乘以：

$$
P(B)
$$

得到：

$$
P(A\cap B)
=
P(A\mid B)P(B)
$$

这句话的人话意思是：

想要 $A$ 和 $B$ 同时发生，可以这样想：

先让 $B$ 发生，再在 $B$ 已经发生的前提下让 $A$ 发生。

同样，也可以从另一个方向看。

条件概率还有：

$$
P(B\mid A)
=
\frac{P(A\cap B)}{P(A)}
$$

两边同时乘以：

$$
P(A)
$$

得到：

$$
P(A\cap B)
=
P(B\mid A)P(A)
$$

所以同一个交集 $A\cap B$，有两种写法：

$$
P(A\cap B)
=
P(A\mid B)P(B)
$$

以及：

$$
P(A\cap B)
=
P(B\mid A)P(A)
$$

## 3. 贝叶斯公式从哪里来？

上一节得到：

$$
P(A\cap B)
=
P(A\mid B)P(B)
$$

也得到：

$$
P(A\cap B)
=
P(B\mid A)P(A)
$$

因为它们都等于同一个东西：

$$
P(A\cap B)
$$

所以可以令它们相等：

$$
P(A\mid B)P(B)
=
P(B\mid A)P(A)
$$

现在我们想求的是：

$$
P(A\mid B)
$$

于是两边同时除以：

$$
P(B)
$$

得到：

$$
P(A\mid B)
=
\frac{P(B\mid A)P(A)}{P(B)}
$$

这就是贝叶斯公式。

它不是背出来的。

它就是从条件概率公式变形来的。

## 4. 贝叶斯公式每一项是什么意思？

贝叶斯公式是：

$$
P(A\mid B)
=
\frac{P(B\mid A)P(A)}{P(B)}
$$

四个部分分别是：

$$
P(A)
$$

先验概率。

意思是：还没看到 $B$ 这个信息之前，$A$ 本来有多可能发生。

$$
P(B\mid A)
$$

似然。

意思是：如果 $A$ 真的发生，那么观察到 $B$ 的可能性有多大。

$$
P(B)
$$

证据。

意思是：$B$ 这个现象本身出现的概率有多大。

$$
P(A\mid B)
$$

后验概率。

意思是：已经看到 $B$ 以后，我们重新判断 $A$ 的概率。

人话版：

$$
后验概率
=
\frac{似然\times 先验概率}{证据}
$$

更人话一点：

原来我对 $A$ 有一个初始判断。

现在我看到了新证据 $B$。

这个证据如果在 $A$ 成立时很容易出现，那 $A$ 的可信度就上升。

如果这个证据在 $A$ 成立时也不太容易出现，那 $A$ 的可信度就不会上升太多。

## 5. 一个具体例子：红色水果是不是苹果？

假设桌上有 $100$ 个水果。

其中苹果有 $60$ 个，香蕉有 $40$ 个。

苹果里有 $48$ 个是红色的。

香蕉里有 $8$ 个是红色的。

现在拿到一个水果，只知道它是红色。

问：

它是苹果的概率是多少？

设：

$$
A=苹果
$$

$$
R=红色
$$

我们要算的是：

$$
P(A\mid R)
$$

也就是：

已经知道它是红色的条件下，它是苹果的概率。

## 6. 先用数数的方法算

红色苹果有：

$$
48
$$

红色香蕉有：

$$
8
$$

所以红色水果总共有：

$$
48+8=56
$$

在这 $56$ 个红色水果里，苹果有 $48$ 个。

所以：

$$
P(A\mid R)
=
\frac{48}{56}
=
\frac{6}{7}
\approx
0.8571
$$

人话版：

不是问所有水果里苹果占多少。

而是先缩小到“红色水果”这个范围。

然后问红色水果里面苹果占多少。

## 7. 再用贝叶斯公式算

贝叶斯公式：

$$
P(A\mid R)
=
\frac{P(R\mid A)P(A)}{P(R)}
$$

先验概率：

$$
P(A)
=
\frac{60}{100}
=
0.6
$$

似然：

$$
P(R\mid A)
=
\frac{48}{60}
=
0.8
$$

证据：

$$
P(R)
=
\frac{56}{100}
=
0.56
$$

代入：

$$
P(A\mid R)
=
\frac{0.8\times 0.6}{0.56}
$$

先算分子：

$$
0.8\times 0.6=0.48
$$

所以：

$$
P(A\mid R)
=
\frac{0.48}{0.56}
=
\frac{6}{7}
\approx
0.8571
$$

和直接数数得到的结果一样。

## 8. 分母 $P(R)$ 到底在干嘛？

很多人第一次看贝叶斯公式，最容易卡在分母：

$$
P(R)
$$

它的作用是归一化。

也就是把所有可能类别对“红色”这个现象的贡献都加起来。

在这个例子里，红色可能来自苹果，也可能来自香蕉。

所以：

$$
P(R)
=
P(R\mid A)P(A)
+
P(R\mid B)P(B)
$$

这里：

$$
B=香蕉
$$

苹果贡献的红色概率是：

$$
P(R\mid A)P(A)
=
0.8\times 0.6
=
0.48
$$

香蕉贡献的红色概率是：

$$
P(R\mid B)P(B)
=
\frac{8}{40}\times 0.4
=
0.2\times 0.4
=
0.08
$$

所以：

$$
P(R)
=
0.48+0.08
=
0.56
$$

人话版：

分母是在问：

红色这个现象，一共有多常见？

分子是在问：

红色并且是苹果，这部分有多少？

所以：

$$
\frac{红色苹果}{全部红色}
$$

就是：

$$
P(苹果\mid 红色)
$$

## 9. 贝叶斯分类器的基本想法

机器学习里，我们通常不是只判断一个类别。

而是有多个类别：

$$
y_1,y_2,\ldots,y_K
$$

看到特征 $x$ 以后，我们想算：

$$
P(y_k\mid x)
$$

贝叶斯公式写成：

$$
P(y_k\mid x)
=
\frac{P(x\mid y_k)P(y_k)}{P(x)}
$$

分类时，我们选择后验概率最大的类别：

$$
\hat{y}
=
\arg\max_{y_k}
P(y_k\mid x)
$$

把贝叶斯公式代进去：

$$
\hat{y}
=
\arg\max_{y_k}
\frac{P(x\mid y_k)P(y_k)}{P(x)}
$$

注意：

对所有类别来说，看到的是同一个 $x$。

所以：

$$
P(x)
$$

对所有类别都一样。

比较谁大时，分母一样，可以不看。

因此分类规则可以写成：

$$
\hat{y}
=
\arg\max_{y_k}
P(x\mid y_k)P(y_k)
$$

人话版：

看某个类别本来常不常见。

再看如果它真是这个类别，出现这些特征的可能性大不大。

两者相乘，谁最大，就选谁。

## 10. 本课小结

这节只需要抓住三件事。

第一，条件概率是在缩小观察范围：

$$
P(A\mid B)
=
\frac{P(A\cap B)}{P(B)}
$$

第二，贝叶斯公式来自条件概率的两种写法：

$$
P(A\cap B)
=
P(A\mid B)P(B)
$$

$$
P(A\cap B)
=
P(B\mid A)P(A)
$$

所以：

$$
P(A\mid B)
=
\frac{P(B\mid A)P(A)}{P(B)}
$$

第三，分类时最关心的是：

$$
P(y\mid x)
$$

也就是：

看到特征 $x$ 以后，样本属于类别 $y$ 的概率。

重点会是：

为什么要假设特征之间条件独立，以及这个假设到底“朴素”在哪里。

```python
apple = 60
banana = 40
red_apple = 48
red_banana = 8

p_apple = apple / 100
p_red_given_apple = red_apple / apple
p_red = (red_apple + red_banana) / 100

p_apple_given_red = p_red_given_apple * p_apple / p_red

print("P(苹果) =", p_apple)
print("P(红色 | 苹果) =", p_red_given_apple)
print("P(红色) =", p_red)
print("P(苹果 | 红色) =", p_apple_given_red)
```
