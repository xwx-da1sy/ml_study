# 贝叶斯第二课：朴素贝叶斯

这一课开始学习朴素贝叶斯。

先说结论：

朴素贝叶斯的核心不是“贝叶斯公式很复杂”。

它真正的核心是一个很大胆的简化：

在已知类别 $y$ 的条件下，特征之间互相独立。

也就是：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
P(x_1\mid y)
P(x_2\mid y)
\cdots
P(x_n\mid y)
$$

这个假设通常不完全真实。

比如一封邮件里出现“中奖”，往往也更容易出现“领取”“链接”。

这些词之间明显不是完全独立的。

但朴素贝叶斯依然常常有用。

原因是：

它不一定要把真实世界建模得完全准确。

它只需要在分类时，把不同类别大致排出正确顺序。

## 1. 从贝叶斯分类器开始

分类问题里，我们看到一个样本：

$$
x
$$

想判断它属于哪个类别：

$$
y
$$

贝叶斯分类器关心的是：

$$
P(y\mid x)
$$

也就是：

看到特征 $x$ 以后，类别是 $y$ 的概率。

根据贝叶斯公式：

$$
P(y\mid x)
=
\frac{P(x\mid y)P(y)}{P(x)}
$$

分类时，我们选择后验概率最大的类别：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(y\mid x)
$$

把贝叶斯公式代进去：

$$
\hat{y}
=
\underset{y}{\arg\max}
\frac{P(x\mid y)P(y)}{P(x)}
$$

注意：

对所有候选类别来说，当前样本 $x$ 是同一个。

所以：

$$
P(x)
$$

对所有类别都是同一个分母。

比较大小时，分母一样，可以去掉。

于是：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(x\mid y)P(y)
$$

人话版：

一个类别最终能不能赢，取决于两件事。

第一：

$$
P(y)
$$

这个类别本来有多常见。

第二：

$$
P(x\mid y)
$$

如果它真的是这个类别，那么出现当前这些特征的可能性有多大。

## 2. 为什么会遇到麻烦？

上面写的是：

$$
P(x\mid y)
$$

但真实样本通常不只有一个特征。

比如一封邮件可以由很多词组成：

$$
x_1=优惠
$$

$$
x_2=中奖
$$

$$
x_3=链接
$$

如果有 $n$ 个特征，样本可以写成：

$$
x=(x_1,x_2,\ldots,x_n)
$$

那么贝叶斯分类器需要算：

$$
P(x_1,x_2,\ldots,x_n\mid y)
$$

这表示：

如果类别是 $y$，那么这些特征同时出现的概率是多少。

麻烦就在“同时”两个字。

如果每个特征都有很多可能取值，那么所有特征组合会非常多。

训练集中很难把每一种组合都见一遍。

比如有 $10$ 个特征，每个特征都有 $5$ 种取值。

所有组合数量是：

$$
5^{10}
$$

这是：

$$
9765625
$$

将近一千万种组合。

如果直接估计：

$$
P(x_1,x_2,\ldots,x_n\mid y)
$$

需要大量数据。

这就是朴素贝叶斯要解决的问题。

## 3. 朴素假设到底是什么？

朴素贝叶斯做了一个简化：

在类别 $y$ 已经确定的前提下，各个特征之间条件独立。

公式写成：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
\prod_{j=1}^{n}
P(x_j\mid y)
$$

如果展开写，就是：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
P(x_1\mid y)
P(x_2\mid y)
\cdots
P(x_n\mid y)
$$

这句话不是说：

特征在任何情况下都独立。

它说的是：

当类别 $y$ 已经确定以后，特征之间近似独立。

比如判断邮件是不是垃圾邮件。

如果已经知道它是垃圾邮件，那么“优惠”“中奖”“链接”这些词的出现，朴素贝叶斯会近似看成彼此独立。

这当然有点粗糙。

所以它叫“朴素”。

人话版：

朴素贝叶斯不去精细研究特征之间怎么互相影响。

它只问：

在某个类别下，每个特征各自出现的可能性是多少。

然后把这些可能性乘起来。

## 4. 朴素贝叶斯公式怎么来？

从贝叶斯分类器开始：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(x\mid y)P(y)
$$

当样本有多个特征：

$$
x=(x_1,x_2,\ldots,x_n)
$$

就写成：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(x_1,x_2,\ldots,x_n\mid y)P(y)
$$

现在使用朴素假设：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
\prod_{j=1}^{n}
P(x_j\mid y)
$$

代进去：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
$$

这就是朴素贝叶斯分类器的核心公式。

说人话：

先看类别本身常不常见。

再看这些特征在这个类别下各自常不常见。

把这些证据合起来，哪个类别分数最大，就预测哪个类别。

## 5. 为什么实际计算常用对数？

朴素贝叶斯要乘很多概率：

$$
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
$$

每个概率都在 $0$ 到 $1$ 之间。

如果特征很多，很多小数连续相乘，结果会变得非常小。

计算机里可能出现数值下溢。

所以实际计算经常取对数。

因为对数有一个性质：

$$
\log(ab)
=
\log a+\log b
$$

所以：

$$
\log\left[
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
\right]
=
\log P(y)
+
\sum_{j=1}^{n}
\log P(x_j\mid y)
$$

分类时，取对数不会改变大小顺序。

因此可以写成：

$$
\hat{y}
=
\underset{y}{\arg\max}
\left[
\log P(y)
+
\sum_{j=1}^{n}
\log P(x_j\mid y)
\right]
$$

人话版：

乘法容易把数乘得太小。

取对数以后，乘法变加法，更稳定，也更好算。

## 6. 手算例子：垃圾邮件分类

现在做一个小例子。

我们只看四个词：

$$
优惠,\ 中奖,\ 会议,\ 项目
$$

词表大小是：

$$
K=4
$$

类别有两个：

$$
y=垃圾邮件
$$

和：

$$
y=正常邮件
$$

训练集中一共有 $10$ 封邮件。

其中垃圾邮件 $4$ 封，正常邮件 $6$ 封。

所以先验概率是：

$$
P(垃圾邮件)
=
\frac{4}{10}
=
0.4
$$

$$
P(正常邮件)
=
\frac{6}{10}
=
0.6
$$

现在统计词频。

在垃圾邮件中：

| 词 | 出现次数 |
|---|---:|
| 优惠 | $6$ |
| 中奖 | $4$ |
| 会议 | $0$ |
| 项目 | $1$ |

垃圾邮件中这四个词总共出现：

$$
6+4+0+1=11
$$

在正常邮件中：

| 词 | 出现次数 |
|---|---:|
| 优惠 | $1$ |
| 中奖 | $0$ |
| 会议 | $5$ |
| 项目 | $4$ |

正常邮件中这四个词总共出现：

$$
1+0+5+4=10
$$

现在有一封新邮件，里面只看到了两个词：

$$
优惠,\ 中奖
$$

我们要判断它更像垃圾邮件，还是正常邮件。

## 7. 不加平滑会发生什么？

如果不加平滑，正常邮件中“中奖”出现次数是：

$$
0
$$

那么：

$$
P(中奖\mid 正常邮件)
=
\frac{0}{10}
=
0
$$

于是正常邮件的分数会包含：

$$
P(中奖\mid 正常邮件)
$$

也就是包含一个：

$$
0
$$

所以整个正常邮件分数会变成：

$$
0
$$

这不太合理。

因为训练集中没见过，不代表现实中正常邮件绝对不会出现“中奖”这个词。

所以需要拉普拉斯平滑。

## 8. 加入拉普拉斯平滑

使用平滑系数：

$$
\alpha=1
$$

多项式朴素贝叶斯里，词概率估计为：

$$
P(w\mid y)
=
\frac{N_{yw}+\alpha}{N_y+\alpha K}
$$

这里：

- $N_{yw}$：类别 $y$ 下，词 $w$ 出现的次数。
- $N_y$：类别 $y$ 下，词表内所有词出现的总次数。
- $K$：词表大小。
- $\alpha$：平滑系数。

先算垃圾邮件。

垃圾邮件总词数：

$$
N_{垃圾}=11
$$

分母是：

$$
N_{垃圾}+\alpha K
=
11+1\times 4
=
15
$$

所以：

$$
P(优惠\mid 垃圾邮件)
=
\frac{6+1}{15}
=
\frac{7}{15}
$$

$$
P(中奖\mid 垃圾邮件)
=
\frac{4+1}{15}
=
\frac{5}{15}
$$

再算正常邮件。

正常邮件总词数：

$$
N_{正常}=10
$$

分母是：

$$
N_{正常}+\alpha K
=
10+1\times 4
=
14
$$

所以：

$$
P(优惠\mid 正常邮件)
=
\frac{1+1}{14}
=
\frac{2}{14}
$$

$$
P(中奖\mid 正常邮件)
=
\frac{0+1}{14}
=
\frac{1}{14}
$$

你看，原来为 $0$ 的概率现在变成了：

$$
\frac{1}{14}
$$

这就是拉普拉斯平滑的作用。

## 9. 计算两个类别的分数

新邮件包含：

$$
优惠,\ 中奖
$$

垃圾邮件分数：

$$
Score(垃圾)
=
P(垃圾)
P(优惠\mid 垃圾)
P(中奖\mid 垃圾)
$$

代入：

$$
Score(垃圾)
=
0.4
\times
\frac{7}{15}
\times
\frac{5}{15}
$$

先算后两项：

$$
\frac{7}{15}
\times
\frac{5}{15}
=
\frac{35}{225}
$$

所以：

$$
Score(垃圾)
=
0.4
\times
\frac{35}{225}
$$

$$
Score(垃圾)
\approx
0.0622
$$

正常邮件分数：

$$
Score(正常)
=
P(正常)
P(优惠\mid 正常)
P(中奖\mid 正常)
$$

代入：

$$
Score(正常)
=
0.6
\times
\frac{2}{14}
\times
\frac{1}{14}
$$

先算后两项：

$$
\frac{2}{14}
\times
\frac{1}{14}
=
\frac{2}{196}
$$

所以：

$$
Score(正常)
=
0.6
\times
\frac{2}{196}
$$

$$
Score(正常)
\approx
0.0061
$$

因为：

$$
Score(垃圾)>Score(正常)
$$

所以预测为：

$$
垃圾邮件
$$

## 10. 这里为什么没有除以 $P(x)$？

完整贝叶斯公式是：

$$
P(y\mid x)
=
\frac{P(x\mid y)P(y)}{P(x)}
$$

但分类时，我们比较的是不同类别。

对于同一封新邮件：

$$
x=优惠,\ 中奖
$$

无论你在比较垃圾邮件还是正常邮件，分母都是同一个：

$$
P(x)
$$

所以：

$$
\frac{P(x\mid 垃圾)P(垃圾)}{P(x)}
$$

和：

$$
\frac{P(x\mid 正常)P(正常)}{P(x)}
$$

比较大小时，分母一样。

可以只比较分子：

$$
P(x\mid y)P(y)
$$

人话版：

我们不是要得到精确概率值。

我们只是要知道哪个类别更大。

所以共同分母可以先不算。

## 11. 多项式、伯努利、高斯朴素贝叶斯

朴素贝叶斯不是只有一种。

常见有三种。

### 多项式朴素贝叶斯

适合词频。

比如文本分类中，一个词出现了几次。

常用公式：

$$
P(w\mid y)
=
\frac{N_{yw}+\alpha}{N_y+\alpha K}
$$

这里关心的是词出现的次数。

### 伯努利朴素贝叶斯

适合二值特征。

比如某个词是否出现。

只关心：

$$
出现
$$

或者：

$$
不出现
$$

而不关心出现了几次。

### 高斯朴素贝叶斯

适合连续特征。

比如身高、体重、血糖、花瓣长度。

它假设在某个类别下，某个连续特征服从高斯分布：

$$
x_j\mid y
\sim
N(\mu_{yj},\sigma_{yj}^2)
$$

然后用高斯密度函数估计：

$$
P(x_j\mid y)
$$

这三种方法的共同点是：

都使用朴素假设：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
\prod_{j=1}^{n}
P(x_j\mid y)
$$

区别只是每个：

$$
P(x_j\mid y)
$$

怎么估计。

## 12. 本课小结

朴素贝叶斯的主线是：

第一，从贝叶斯分类器开始：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(x\mid y)P(y)
$$

第二，当样本有多个特征时，直接估计联合概率很难：

$$
P(x_1,x_2,\ldots,x_n\mid y)
$$

第三，朴素贝叶斯做条件独立假设：

$$
P(x_1,x_2,\ldots,x_n\mid y)
=
\prod_{j=1}^{n}
P(x_j\mid y)
$$

第四，所以分类规则变成：

$$
\hat{y}
=
\underset{y}{\arg\max}
P(y)
\prod_{j=1}^{n}
P(x_j\mid y)
$$

第五，拉普拉斯平滑用于避免概率为 $0$：

$$
P(x=k\mid y)
=
\frac{N_{yk}+\alpha}{N_y+\alpha K}
$$

一句话：

朴素贝叶斯把复杂的“多个特征同时出现的概率”，拆成了多个简单的“单个特征在某类别下出现的概率”。

```python
from fractions import Fraction
import math

classes = {
    "垃圾邮件": {
        "doc_count": 4,
        "word_counts": {"优惠": 6, "中奖": 4, "会议": 0, "项目": 1},
    },
    "正常邮件": {
        "doc_count": 6,
        "word_counts": {"优惠": 1, "中奖": 0, "会议": 5, "项目": 4},
    },
}

new_email_words = ["优惠", "中奖"]
alpha = 1
vocab = ["优惠", "中奖", "会议", "项目"]
K = len(vocab)
total_docs = sum(info["doc_count"] for info in classes.values())

scores = {}
details = {}

for label, info in classes.items():
    prior = info["doc_count"] / total_docs
    total_words = sum(info["word_counts"].values())
    score = prior
    log_score = math.log(prior)
    probs = {}

    for word in new_email_words:
        prob = (info["word_counts"][word] + alpha) / (total_words + alpha * K)
        probs[word] = prob
        score *= prob
        log_score += math.log(prob)

    scores[label] = score
    details[label] = {
        "prior": prior,
        "total_words": total_words,
        "word_probs": probs,
        "score": score,
        "log_score": log_score,
    }

for label, info in details.items():
    print(label)
    print("  先验概率:", info["prior"])
    print("  类别总词数:", info["total_words"])
    print("  词概率:", info["word_probs"])
    print("  原始分数:", info["score"])
    print("  对数分数:", info["log_score"])

prediction = max(scores, key=scores.get)
print("预测类别:", prediction)
```
