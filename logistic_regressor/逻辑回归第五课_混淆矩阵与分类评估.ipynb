## 1. 为什么要学混淆矩阵

在分类问题里，模型最后给出的不是连续值，而是类别判断。

这时只看一个总的准确率往往不够，因为它没有告诉我们：

- 模型把多少正类判对了
- 模型把多少负类判对了
- 模型更容易错在哪一类上

混淆矩阵就是把这些信息拆开来看。

## 2. 混淆矩阵长什么样

这很重要，对于二分类问题，混淆矩阵通常写成：

$$
\begin{array}{c|cc}
 & \text{预测为正类} & \text{预测为负类} \\
\hline
\text{真实为正类} & TP & FN \\
\text{真实为负类} & FP & TN
\end{array}
$$

这里四个缩写的含义是：

- $TP$：True Positive，真实是正类，预测也是正类
- $FN$：False Negative，真实是正类，但预测成了负类
- $FP$：False Positive，真实是负类，但预测成了正类
- $TN$：True Negative，真实是负类，预测也是负类


## 3. 先把四个量讲清楚

如果我们把“1”定义成正类，把“0”定义成负类，那么：

- 真实是 $1$，预测也是 $1$，这是 $TP$
- 真实是 $1$，预测成了 $0$，这是 $FN$
- 真实是 $0$，预测成了 $1$，这是 $FP$
- 真实是 $0$，预测也是 $0$，这是 $TN$

因此混淆矩阵其实就是把预测结果按这四种情况分开统计。

## 4. 怎样判断 true / false / positive / negative

这个地方最容易混，所以单独拆开说。

先固定约定：

$$
1 = \text{positive}, \qquad 0 = \text{negative}
$$

也就是说：

- 正类就是标签 $1$
- 负类就是标签 $0$

然后记住两条规则：

- positive / negative 看的是“预测成了哪一类”
- true / false 看的是“这个预测对不对”

于是四种情况就能完全确定下来。

### 4.1 True Positive

真实是正类，预测也是正类：

$$
y_{\text{true}} = 1, \qquad y_{\text{pred}} = 1
$$

这叫：

$$
TP
$$

这里：

- positive：因为预测成了正类
- true：因为预测正确

### 4.2 False Positive

真实是负类，但预测成了正类：

$$
y_{\text{true}} = 0, \qquad y_{\text{pred}} = 1
$$

这叫：

$$
FP
$$

这里：

- positive：因为预测成了正类
- false：因为预测错了

### 4.3 False Negative

真实是正类，但预测成了负类：

$$
y_{\text{true}} = 1, \qquad y_{\text{pred}} = 0
$$

这叫：

$$
FN
$$

这里：

- negative：因为预测成了负类
- false：因为预测错了

### 4.4 True Negative

真实是负类，预测也是负类：

$$
y_{\text{true}} = 0, \qquad y_{\text{pred}} = 0
$$

这叫：

$$
TN
$$

这里：

- negative：因为预测成了负类
- true：因为预测正确

所以可以压缩成一张表：

$$
\begin{array}{c|c|c}
y_{\text{true}} & y_{\text{pred}} & \text{名称} \\
\hline
1 & 1 & TP \\
0 & 1 & FP \\
1 & 0 & FN \\
0 & 0 & TN
\end{array}
$$

最实用的记忆方法是：

- 先看预测成了正类还是负类
- 再看这个预测是对还是错

也就是说：

- 预测成正类：要么是 $TP$，要么是 $FP$
- 预测成负类：要么是 $TN$，要么是 $FN$
- 预测对了：是 true
- 预测错了：是 false

## 5. 一个最小例子

假设：

$$
y_{\text{true}} = [1, 1, 1, 0, 0, 0, 1, 0]
$$

$$
y_{\text{pred}} = [1, 0, 1, 0, 1, 0, 1, 0]
$$

逐个对照：

- 第 1 个：真实 1，预测 1，属于 $TP$
- 第 2 个：真实 1，预测 0，属于 $FN$
- 第 3 个：真实 1，预测 1，属于 $TP$
- 第 4 个：真实 0，预测 0，属于 $TN$
- 第 5 个：真实 0，预测 1，属于 $FP$
- 第 6 个：真实 0，预测 0，属于 $TN$
- 第 7 个：真实 1，预测 1，属于 $TP$
- 第 8 个：真实 0，预测 0，属于 $TN$

于是统计结果是：

$$
TP = 3, \qquad FN = 1, \qquad FP = 1, \qquad TN = 3
$$


## 6. 由混淆矩阵推出常见指标

### 5.1 准确率 accuracy

准确率表示：所有样本中，预测正确的比例。

$$
\mathrm{accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

### 5.2 精确率 precision

精确率表示：在所有“被预测成正类”的样本中，真正是正类的比例。

$$
\mathrm{precision} = \frac{TP}{TP + FP}
$$

### 5.3 召回率 recall

召回率表示：在所有“真实为正类”的样本中，被成功找回来的比例。

$$
\mathrm{recall} = \frac{TP}{TP + FN}
$$

### 5.4 $F_1$ 分数

$F_1$ 是 precision 和 recall 的调和平均数：

$$
F_1 = \frac{2 \cdot \mathrm{precision} \cdot \mathrm{recall}}{\mathrm{precision} + \mathrm{recall}}
$$

把公式展开后，也常写成：

$$
F_1 = \frac{2TP}{2TP + FP + FN}
$$


## 7. 用上面的例子手算一遍

刚才我们得到：

$$
TP = 3, \qquad TN = 3, \qquad FP = 1, \qquad FN = 1
$$

所以：

$$
\mathrm{accuracy} = \frac{3+3}{3+3+1+1} = \frac{6}{8} = 0.75
$$

$$
\mathrm{precision} = \frac{3}{3+1} = \frac{3}{4} = 0.75
$$

$$
\mathrm{recall} = \frac{3}{3+1} = \frac{3}{4} = 0.75
$$

$$
F_1 = \frac{2 \cdot 0.75 \cdot 0.75}{0.75 + 0.75} = 0.75
$$

这个例子刚好比较对称，所以四个指标最后都等于 $0.75$。

```python
import numpy as np
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

y_true = np.array([1, 1, 1, 0, 0, 0, 1, 0])
y_pred = np.array([1, 0, 1, 0, 1, 0, 1, 0])

cm = confusion_matrix(y_true, y_pred)
acc = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print('confusion matrix =')
print(cm)
print('accuracy =', acc)
print('precision =', precision)
print('recall =', recall)
print('f1 =', f1)
```

## 8. 为什么只看准确率不够

假设一个数据集里有 100 条样本，其中：

- 95 条是负类
- 5 条是正类

如果一个模型什么都不学，永远都预测成负类，那么它的准确率就是：

$$
\mathrm{accuracy} = \frac{95}{100} = 95\%
$$

看起来很高，但这个模型实际上一个正类也没找出来。

这时：

$$
TP = 0, \qquad FN = 5
$$

于是 recall 会非常差。

这就是为什么在分类问题里，混淆矩阵比单独看准确率更重要。

## 9. 混淆矩阵和逻辑回归有什么关系

逻辑回归先输出的是概率：

$$
p = P(y=1 \mid x)
$$

然后再根据阈值把概率变成类别：

- 如果 $p \ge 0.5$，预测为正类
- 如果 $p < 0.5$，预测为负类

一旦变成了类别预测，就可以用混淆矩阵去评估。

也就是说：

- 交叉熵是在训练阶段衡量概率预测好不好
- 混淆矩阵是在评估阶段衡量分类结果好不好

\hline
\text{真实正类} & TP & FN \\
\text{真实负类} & FP & TN
\end{array}
$$

$$
\mathrm{accuracy} = \frac{TP + TN}{TP + TN + FP + FN}
$$

$$
\mathrm{precision} = \frac{TP}{TP + FP}
$$

$$
\mathrm{recall} = \frac{TP}{TP + FN}
$$

$$
F_1 = \frac{2TP}{2TP + FP + FN}
$$
