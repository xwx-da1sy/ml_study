# 逻辑回归第一课：数学基础

这份 Notebook 用来整理进入逻辑回归之前最需要掌握的数学基础，重点放在：

- 概率与条件概率
- 伯努利分布与二分类标签
- odds 与 log-odds
- sigmoid 函数
- 从线性分数到分类概率
- 决策边界为什么仍然是线性的


## 1. 为什么要先学这一节

在线性回归中，我们预测的是一个连续值：

$$
\hat{y} = Xw + b
$$

但在逻辑回归中，我们通常面对的是二分类问题，标签只取两种值：

$$
y \in \{0,1\}
$$

于是问题变成：给定特征 $x$，我们如何预测样本属于 1 类的概率？

$$
P(y=1 \mid x)
$$

所以逻辑回归真正的数学入口不是先看代码，而是先理解概率、条件概率、概率分布，以及概率怎样和线性模型连起来。

## 2. 基本记号

为了和前面的线性回归笔记保持一致，我们继续使用下面这些记号：

$$
X \in \mathbb{R}^{m \times n}
$$

其中：

- $m$ 是样本数
- $n$ 是特征数

第 $i$ 个样本记作：

$$
x^{(i)} \in \mathbb{R}^{n}
$$

标签向量记作：

$$
y \in \{0,1\}^{m}
$$

参数记作：

$$
w \in \mathbb{R}^{n}, \qquad b \in \mathbb{R}
$$

线性分数写成：

$$
z = w^T x + b
$$


## 3. 概率与条件概率

概率可以理解成一件事情发生的可能性大小。若事件 $A$ 的概率记作 $P(A)$，则有：

$$
0 \le P(A) \le 1
$$

逻辑回归里最重要的不是一般概率，而是条件概率：

$$
P(A \mid B)
$$

意思是：在事件 $B$ 已经发生的条件下，事件 $A$ 发生的概率。

在机器学习里，通常把它理解为：给定输入特征 $x$，标签属于某一类的概率。

因此逻辑回归学习的不是：

$$
P(y=1)
$$

而是：

$$
P(y=1 \mid x)
$$


## 4. 正类、负类与伯努利分布

在二分类问题里，我们人为规定：

- $y=1$ 叫正类
- $y=0$ 叫负类

这里的“正类”和“负类”只是标签命名，不带好坏含义。

逻辑回归默认建模的是正类概率：

$$
p = P(y=1 \mid x)
$$

因此负类概率就是：

$$
P(y=0 \mid x) = 1-p
$$

由于标签只有 0 和 1 两种取值，这种随机变量适合用伯努利分布描述：

$$
Y \sim \mathrm{Bernoulli}(p)
$$

也就是说：

$$
P(Y=1)=p, \qquad P(Y=0)=1-p
$$


## 5. 为什么单个样本的概率能统一写成一个公式

对于单个样本，若真实标签 $y$ 只能取 0 或 1，那么：

- 当 $y=1$ 时，对应概率应该是 $p$
- 当 $y=0$ 时，对应概率应该是 $1-p$

这两种情况可以统一写成：

$$
P(y \mid x) = p^y (1-p)^{1-y}
$$

验证一下：

当 $y=1$ 时：

$$
p^1(1-p)^0 = p
$$

当 $y=0$ 时：

$$
p^0(1-p)^1 = 1-p
$$

这个公式后面会直接连到交叉熵。

## 6. odds 与 log-odds

如果一个事件发生的概率是 $p$，不发生的概率是 $1-p$，那么它的 odds 定义为：

$$
\mathrm{odds} = \frac{p}{1-p}
$$

再对 odds 取对数，就得到 log-odds：

$$
\log \frac{p}{1-p}
$$

它的重要性在于：

- 概率 $p$ 的范围是 $(0,1)$
- odds 的范围是 $(0,+\infty)$
- log-odds 的范围是 $(-\infty,+\infty)$

而线性函数 $w^T x + b$ 也可以取任意实数，因此逻辑回归的核心假设之一就是：

$$
\log \frac{p}{1-p} = w^T x + b
$$


## 7. 从 log-odds 推出 sigmoid

设：

$$
z = w^T x + b
$$

并令：

$$
\log \frac{p}{1-p} = z
$$

则有：

$$
\frac{p}{1-p} = e^z
$$

整理得到：

$$
p = \frac{e^z}{1+e^z} = \frac{1}{1+e^{-z}}
$$

于是得到 sigmoid 函数：

$$
\sigma(z) = \frac{1}{1+e^{-z}}
$$

逻辑回归不是随便选了一个函数，而是从 log-odds 的线性假设自然推出了 sigmoid。

## 8. sigmoid 函数的性质

sigmoid 函数：

$$
\sigma(z) = \frac{1}{1+e^{-z}}
$$

有几个重要性质：

- $0 < \sigma(z) < 1$
- 当 $z=0$ 时，$\sigma(z)=0.5$
- 当 $z \to +\infty$ 时，$\sigma(z) \to 1$
- 当 $z \to -\infty$ 时，$\sigma(z) \to 0$
- $\sigma(z)$ 是单调递增函数

这使得它非常适合拿来表示概率。

## 9. 从线性分数到分类概率

逻辑回归最核心的预测过程可以写成两步：

第一步，先算线性分数：

$$
z = w^T x + b
$$

第二步，再把它映射成概率：

$$
p = \sigma(z) = \frac{1}{1+e^{-z}}
$$

并把这个概率解释为：

$$
p = P(y=1 \mid x)
$$

于是也就有：

$$
P(y=0 \mid x) = 1-p
$$


## 10. 阈值与决策边界

逻辑回归先输出概率，最终再根据阈值给出类别。

最常见的阈值是 $0.5$：

- 如果 $p \ge 0.5$，预测为正类
- 如果 $p < 0.5$，预测为负类

由于：

$$
\sigma(0) = 0.5
$$

所以分类边界满足：

$$
p = 0.5 \iff z = 0
$$

再代入 $z = w^T x + b$，得到：

$$
w^T x + b = 0
$$


## 11. 为什么逻辑回归的决策边界仍然是线性的

逻辑回归虽然使用了非线性的 sigmoid 函数，但真正决定分类边界的位置的是线性分数 $w^T x + b$。

因为 sigmoid 是单调递增函数，所以：

$$
p = 0.5 \iff z = 0 \iff w^T x + b = 0
$$

而 $w^T x + b = 0$ 是关于输入变量 $x$ 的线性方程，因此决策边界仍然是线性的。

更一般地，如果阈值不是 $0.5$，而是任意 $t \in (0,1)$，那么边界满足：

$$
\sigma(z) = t
$$

整理可得：

$$
z = \ln\frac{t}{1-t}
$$

于是边界变成：

$$
w^T x + b = \ln\frac{t}{1-t}
$$

右边只是常数，所以它依然是线性的。

## 12. 本节小结

这一节可以压缩成下面这条主线：

$$
y \in \{0,1\}
$$

$$
p = P(y=1 \mid x)
$$

$$
\log\frac{p}{1-p} = w^T x + b
$$

$$
p = \sigma(w^T x + b) = \frac{1}{1+e^{-(w^T x + b)}}
$$

下一步自然会进入交叉熵，也就是逻辑回归中最重要的损失函数。

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z_values = np.array([-3, -1, 0, 1, 3])
p_values = sigmoid(z_values)

for z, p in zip(z_values, p_values):
    print(f"z = {z:>2}, sigmoid(z) = {p:.6f}")
```
