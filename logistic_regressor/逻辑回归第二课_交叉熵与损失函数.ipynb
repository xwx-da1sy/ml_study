## 1. 先说明目标

逻辑回归输出的是正类概率：

$$
p = P(y=1 \mid x)
$$

因此负类概率就是：

$$
P(y=0 \mid x) = 1-p
$$

对一个样本来说，模型越应该把更大的概率分配给真实标签。

交叉熵就是把这件事写成一个可以优化的损失函数。

## 2. 为什么逻辑回归通常不用 MSE

在线性回归里，我们常用均方误差：

$$
J_{\mathrm{MSE}} = \frac{1}{m}\sum_{i=1}^{m}(y^{(i)} - \hat{y}^{(i)})^2
$$

如果把逻辑回归中的预测概率记作 $p^{(i)}$，形式上当然也可以写成：

$$
J = \frac{1}{m}\sum_{i=1}^{m}(y^{(i)} - p^{(i)})^2
$$

但逻辑回归是一个概率模型，它的标签天然对应伯努利分布，因此更自然的做法是从“真实标签的概率应该尽可能大”出发，得到交叉熵。

所以这里不是说 MSE 完全不能写，而是说交叉熵和逻辑回归的概率解释更匹配。

## 3. 单个样本的概率从哪里来

对于单个样本，由于 $y \in \{0,1\}$，它的条件概率可以统一写成：

$$
P(y \mid x) = p^y (1-p)^{1-y}
$$

这个公式的含义是：

- 当 $y=1$ 时，概率变成 $p$
- 当 $y=0$ 时，概率变成 $1-p$

也就是说，这个式子表达的是“模型给真实标签分配的概率”。

## 4. 为什么要最大化这个概率

如果一条样本的真实标签是 1，那么模型给它的 $p$ 越大越合理。

如果一条样本的真实标签是 0，那么模型给它的 $p$ 越小越合理，因为这等价于 $1-p$ 越大。

因此对于单个样本，我们希望：

$$
P(y \mid x)
$$

尽可能大。

对于整个训练集，若样本在给定参数下条件独立，则整个数据集的似然函数写成：

$$
L(w,b) = \prod_{i=1}^{m} P(y^{(i)} \mid x^{(i)})
$$

代入单样本公式得到：

$$
L(w,b) = \prod_{i=1}^{m} \left(p^{(i)}\right)^{y^{(i)}} \left(1-p^{(i)}\right)^{1-y^{(i)}}
$$


## 5. 为什么要取对数

上面的似然函数是一个连乘式，不方便求导和优化。

因此通常对它取对数，把乘积变成求和：

$$
\log L(w,b) = \sum_{i=1}^{m} \log P(y^{(i)} \mid x^{(i)})
$$

再代入：

$$
P(y^{(i)} \mid x^{(i)}) = \left(p^{(i)}\right)^{y^{(i)}} \left(1-p^{(i)}\right)^{1-y^{(i)}}
$$

得到：

$$
\log L(w,b) = \sum_{i=1}^{m} \left[ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) \right]
$$


## 6. 从对数似然到交叉熵

机器学习里更常见的写法是最小化损失，而不是最大化对数似然。

因此给上式加一个负号，就得到负对数似然：

$$
J(w,b) = -\sum_{i=1}^{m} \left[ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) \right]
$$

再除以样本数 $m$，得到平均损失：

$$
J_{\mathrm{CE}}(w,b) = -\frac{1}{m}\sum_{i=1}^{m} [ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) ]
$$

这就是二元交叉熵。

这条公式已经修正为可正常渲染的写法，不会再出现红字高亮。 

## 7. 单个样本时交叉熵长什么样

对于单个样本，交叉熵可以写成：

$$
L(y,p) = -[ y\log p + (1-y)\log(1-p) ]
$$

当 $y=1$ 时：

$$
L(1,p) = -\log p
$$

当 $y=0$ 时：

$$
L(0,p) = -\log(1-p)
$$

这表示：

- 如果真实标签是 1，那么希望 $p$ 尽量大
- 如果真实标签是 0，那么希望 $p$ 尽量小


## 8. 为什么交叉熵符合直觉

看两个例子：

如果真实标签是 1：

$$
-\log(0.9) \approx 0.105
$$

$$
-\log(0.1) \approx 2.303
$$

说明：真实标签是 1 时，模型给出 $0.9$ 比给出 $0.1$ 合理得多。

如果真实标签是 0：

$$
-\log(1-0.1) = -\log(0.9) \approx 0.105
$$

$$
-\log(1-0.9) = -\log(0.1) \approx 2.303
$$

说明：真实标签是 0 时，模型给出 $0.1$ 比给出 $0.9$ 合理得多。

这就是为什么交叉熵特别擅长惩罚“非常自信但判断错了”的预测。

## 9. 交叉熵与 MSE 的直观对比

MSE 更像是在比较“预测值和真实值差了多少”：

$$
J_{\mathrm{MSE}} = \frac{1}{m}\sum_{i=1}^{m}(y^{(i)} - p^{(i)})^2
$$

交叉熵更像是在比较“你给真实类别的概率到底够不够高”：

$$
J_{\mathrm{CE}}(w,b) = -\frac{1}{m}\sum_{i=1}^{m} [ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) ]
$$

对于逻辑回归这种概率模型，后一种衡量方式更自然。

## 10. 本节小结

交叉熵这一节可以压缩成下面这条主线：

$$
P(y \mid x) = p^y (1-p)^{1-y}
$$

$$
\log L(w,b) = \sum_{i=1}^{m} [ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) ]
$$

$$
J_{\mathrm{CE}}(w,b) = -\frac{1}{m}\sum_{i=1}^{m} [ y^{(i)}\log p^{(i)} + (1-y^{(i)})\log(1-p^{(i)}) ]
$$

下一步自然会进入交叉熵对参数的梯度，以及梯度下降如何训练逻辑回归。

```python
import numpy as np

def binary_cross_entropy(y, p):
    eps = 1e-12
    p = np.clip(p, eps, 1 - eps)
    return -(y * np.log(p) + (1 - y) * np.log(1 - p))

examples = [
    (1, 0.9),
    (1, 0.6),
    (1, 0.1),
    (0, 0.1),
    (0, 0.4),
    (0, 0.9),
]

for y, p in examples:
    loss = binary_cross_entropy(y, p)
    print(f"y={y}, p={p:.1f}, loss={loss:.6f}")
```
