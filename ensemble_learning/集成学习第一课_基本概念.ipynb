# 集成学习第一课：基本概念

## 本课目标

理解集成学习为什么有用，以及 Bagging、Boosting、Stacking 三类方法分别在做什么。

一句话版：

```text
集成学习 = 多个模型一起做预测，再把结果合成一个最终答案。
```

## 1. 为什么需要集成学习？

前面已经学过很多单模型：

- KNN
- 线性回归
- 逻辑回归
- 决策树

单模型的问题是：它可能受自身结构限制，也可能受训练数据扰动影响。

例如单棵决策树：

- 优点：规则清楚，容易解释。
- 缺点：容易过拟合，对数据变化敏感。

随机森林的做法是：

```text
不要只相信一棵树，而是训练很多棵树，让它们一起投票。
```

这就是集成学习的核心直觉。

## 2. 什么是基学习器？

集成学习里的每一个单独模型，通常叫基学习器，也可以叫 base learner。

例如随机森林中：

```text
一棵决策树 = 一个基学习器
很多棵决策树 = 一个集成模型
```

基学习器可以是同一种模型，也可以是不同模型。

同一种模型的例子：

```text
100 棵决策树 -> 随机森林
```

不同模型的例子：

```text
逻辑回归 + KNN + 决策树 -> 投票分类器
```

## 3. 分类任务怎么组合？

分类问题中，最直观的组合方式是投票。

假设 5 个模型预测某封邮件是否为垃圾邮件：

| 模型 | 预测 |
|---:|---|
| 模型 1 | 垃圾邮件 |
| 模型 2 | 正常邮件 |
| 模型 3 | 垃圾邮件 |
| 模型 4 | 垃圾邮件 |
| 模型 5 | 正常邮件 |

投票结果：

```text
垃圾邮件：3 票
正常邮件：2 票
```

最终预测：

```text
垃圾邮件
```

````python
# 多数投票的小例子
from collections import Counter

predictions = ["垃圾邮件", "正常邮件", "垃圾邮件", "垃圾邮件", "正常邮件"]
vote_result = Counter(predictions)
final_prediction = vote_result.most_common(1)[0][0]

vote_result, final_prediction
````

## 4. 回归任务怎么组合？

回归问题中，常见方式是取平均。

假设 3 个模型预测一套房子的价格：

| 模型 | 预测房价 |
|---:|---:|
| 模型 1 | 98 万 |
| 模型 2 | 102 万 |
| 模型 3 | 100 万 |

最终预测可以取平均：

```text
(98 + 102 + 100) / 3 = 100 万
```

````python
# 回归平均的小例子
price_predictions = [98, 102, 100]
final_price = sum(price_predictions) / len(price_predictions)

final_price
````

## 5. 集成学习的三大类型

### Bagging

Bagging 的特点是：多个模型并行训练。

每个模型看到的数据不完全一样，最后再投票或平均。

随机森林就是 Bagging 的代表：

```text
训练集
├── 随机抽样 -> 决策树 1
├── 随机抽样 -> 决策树 2
├── 随机抽样 -> 决策树 3
└── 随机抽样 -> 决策树 n

预测时：投票 / 平均
```

Bagging 主要用来降低方差，让模型更稳定。

### Boosting

Boosting 的特点是：多个模型串行训练。

后一个模型会重点处理前一个模型做错或没学好的部分。

可以把它理解成做错题本：

```text
第 1 个模型先做题
-> 找出错题
-> 第 2 个模型重点学错题
-> 再找新的错题
-> 第 3 个模型继续补短板
```

Boosting 的代表算法包括：

- AdaBoost
- GBDT
- XGBoost
- LightGBM

Boosting 常用于提升模型表达能力，但训练过度时也可能过拟合。

### Stacking

Stacking 的特点是：先让多个模型分别预测，再训练一个新模型来学习如何组合这些预测。

直观结构：

```text
原始特征 X
├── 模型 A -> 预测 A
├── 模型 B -> 预测 B
└── 模型 C -> 预测 C

预测 A / B / C -> 第二层模型 -> 最终预测
```

Stacking 更灵活，但实现时要小心数据泄漏。

## 6. 为什么多个模型会更好？

关键不只是“模型数量多”，而是这些模型要有差异。

如果 100 个模型完全一样，那么它们投票的结果和 1 个模型没有区别。

好的集成学习希望做到：

```text
每个模型都有一定能力
+ 模型之间又不完全一样
= 组合后更稳、更准
```

随机森林里的“样本随机”和“特征随机”，本质上就是为了让每棵树不一样。

## 7. 偏差和方差视角

前面学过拟合、欠拟合和泛化。集成学习可以继续从这个角度理解。

Bagging 更常见的作用：

```text
降低方差
```

因为它把多个不稳定模型的结果平均掉，让整体预测不容易被单个模型带偏。

Boosting 更常见的作用：

```text
降低偏差
```

因为它一轮一轮修正错误，让模型逐步逼近更复杂的规律。

注意这不是绝对规则，只是学习初期很好用的理解入口。

## 8. 本课小结

- 集成学习不是一个模型，而是一类“组合多个模型”的方法。
- 分类任务常用投票，回归任务常用平均。
- Bagging 并行训练多个模型，代表算法是随机森林。
- Boosting 串行训练多个模型，后面的模型修正前面的错误。
- Stacking 用第二层模型学习如何组合第一层模型的预测。
- 多个模型要有差异，集成才有意义。

记忆方式：

```text
Bagging：大家并行做题，最后投票。
Boosting：一个接一个改错题。
Stacking：先让大家答题，再训练一个裁判学会怎么参考大家。
```
