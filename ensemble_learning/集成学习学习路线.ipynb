# 集成学习学习路线

## 学习背景

你已经完成或接触过这些基础：

```text
机器学习概述
-> KNN
-> 线性回归
-> 逻辑回归
-> 决策树
-> 随机森林
-> 集成学习（当前）
```

其中“随机森林”其实已经是集成学习的一种典型方法：

```text
随机森林 = 多棵决策树 + 样本随机 + 特征随机 + 投票 / 平均
```

所以这一章不是突然开始一块新大陆，而是从“单个模型”进入“多个模型协作”。

## 1. 本章核心问题

集成学习要回答的问题是：

```text
如果一个模型不够稳定、不够准确，
能不能让多个模型一起工作，得到更好的整体效果？
```

直觉上：

- 单个模型可能会犯错。
- 多个模型如果犯错方式不完全一样，就有机会互相抵消错误。
- 最终组合结果通常比单个模型更稳定。

## 2. 三条主线

### Bagging

Bagging 的重点是“并行训练多个模型，再汇总结果”。

代表算法：

- Bagging
- 随机森林

核心直觉：

```text
让多个模型看到不同数据，降低模型方差，让预测更稳。
```

随机森林就是 Bagging 思想在决策树上的经典应用。

### Boosting

Boosting 的重点是“一个模型接着一个模型训练，后面的模型重点修正前面的错误”。

代表算法：

- AdaBoost
- GBDT
- XGBoost
- LightGBM

核心直觉：

```text
一轮一轮补短板，把多个弱模型叠加成强模型。
```

### Stacking

Stacking 的重点是“用一个新模型学习如何组合多个模型的预测结果”。

核心直觉：

```text
第一层模型负责各自预测，第二层模型负责学习谁更可信。
```

它比投票或平均更灵活，但也更容易因为设计不当而过拟合。

## 3. 建议学习顺序

1. 集成学习基本概念
2. Bagging 与随机森林复盘
3. AdaBoost：从“错题加权”理解 Boosting
4. GBDT：用残差理解梯度提升树
5. XGBoost / LightGBM：工程增强版梯度提升树
6. Voting 与 Stacking：模型组合策略
7. sklearn 实战：分类任务与回归任务
8. 参数调优：树数量、学习率、深度、采样比例

## 4. 本章要反复抓住的两个词

### 方差

方差可以理解为模型对训练数据扰动的敏感程度。

单棵深决策树通常方差高，所以训练数据稍微变一点，树结构和预测结果就可能明显变化。

Bagging 和随机森林主要是在降低方差。

### 偏差

偏差可以理解为模型本身表达能力不足带来的系统性错误。

模型太简单时，训练集和测试集都表现不好，这通常是偏差较高。

Boosting 常常能逐步降低偏差，但如果训练太猛，也可能过拟合。

## 5. 和已有知识的连接

| 已学内容 | 和集成学习的关系 |
|---|---|
| KNN | 可以作为基模型参与投票，但不是最常见主角 |
| 线性回归 | 可以参与回归集成，也能作为 Stacking 的第二层模型 |
| 逻辑回归 | 可作为分类基模型或 Stacking 组合器 |
| 决策树 | 集成学习里最常用的基模型之一 |
| 随机森林 | Bagging 的代表算法 |
| 交叉验证 | 用于评估集成模型和调参 |

## 6. 一句话总览

```text
集成学习 = 让多个模型协作，用组合结果提升稳定性、准确率和泛化能力。
```
