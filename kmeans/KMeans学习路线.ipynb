# K-means 聚类学习路线


## 1. 当前学习位置

```text
机器学习概述
-> KNN
-> 线性回归
-> 逻辑回归
-> 决策树与集成学习
-> 贝叶斯
-> K-means 聚类（当前）
```

前面学习的大部分算法都属于有监督学习：训练数据里同时有特征 `X` 和答案 `y`。

K-means 将开始一条新的主线——无监督学习：

```text
只有特征 X，没有现成标签 y
-> 观察样本之间的距离
-> 自动把相似样本分成若干组
```


## 2. 本章核心问题

> 当数据没有类别标签时，怎样按照样本之间的相似程度，自动发现数据内部的分组？

常见应用包括客户分群、图片颜色压缩、文章分组和探索数据结构。

## 4. 和已有知识的连接

| 已学内容 | 与 K-means 的关系 |
|---|---|
| KNN | 都依赖样本间距离，也都容易受特征尺度影响 |
| 标准化 | 防止数值范围大的特征支配欧氏距离 |
| 线性回归损失 | 都通过一个数值目标衡量当前结果好不好 |
| 决策树 | 树通过特征切分数据，K-means 通过距离划分数据 |
| 交叉验证 | 聚类没有普通分类准确率，需要新的评估方法 |

最直接的知识桥梁是：

```text
KNN：找到离新样本最近的已知样本
K-means：找到离每个样本最近的质心
```

名字中的 `K` 含义并不相同：

```text
KNN 的 K     -> 参考多少个邻居
K-means 的 K -> 把数据分成多少个簇
```
