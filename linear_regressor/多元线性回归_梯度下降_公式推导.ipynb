## 1. 记号说明

设：

$$
X \in \mathbb{R}^{m \times n}
$$

表示特征矩阵，其中：

- $m$ 是样本数
- $n$ 是特征数

设参数向量为：

$$
\theta \in \mathbb{R}^{n}
$$

真实标签向量为：

$$
y \in \mathbb{R}^{m}
$$

如果把偏置项并入设计矩阵，那么也可以写成：

$$
X_{\text{design}} \in \mathbb{R}^{m \times (n+1)}, \qquad
\theta = [b, w_1, w_2, \dots, w_n]^T
$$

后面为了统一记号，我们直接把设计矩阵仍记作 $X$。

## 2. 多元线性回归的预测公式

单个样本的预测值可以写成：

$$
\hat{y}^{(i)} = \theta^T x^{(i)}
$$

把所有样本一起写，就得到矩阵形式：

$$
\hat{y} = X\theta
$$

这里：

$$
\hat{y} \in \mathbb{R}^{m}
$$

表示所有样本的预测值向量。

## 3. 损失函数

如果使用平方误差和（SSE）作为损失函数，那么：

$$
\mathrm{SSE}(\theta) = \sum_{i=1}^{m} \left(y^{(i)} - \hat{y}^{(i)}\right)^2
$$

代入 $\hat{y} = X\theta$，可以写成：

$$
\mathrm{SSE}(\theta) = \sum_{i=1}^{m} \left(y^{(i)} - x^{(i)T}\theta\right)^2
$$

进一步写成向量形式：

$$
\mathrm{SSE}(\theta) = (y - X\theta)^T (y - X\theta)
$$

这也可以写成二范数形式：

$$
\mathrm{SSE}(\theta) = \|y - X\theta\|_2^2
$$

如果使用均方误差风格的写法，则常见形式为：

$$
J(\theta) = \frac{1}{2m} \|y - X\theta\|_2^2
$$

这和 SSE 的最优点相同，只是多了一个常数缩放。

## 4. 残差向量

定义残差向量：

$$
r = y - X\theta
$$

那么损失函数就可以写成：

$$
\mathrm{SSE}(\theta) = r^T r
$$

这表示：

$$
r^T r = r_1^2 + r_2^2 + \cdots + r_m^2
$$

也就是所有样本残差平方的总和。

## 5. 梯度推导

从：

$$
\mathrm{SSE}(\theta) = (y - X\theta)^T (y - X\theta)
$$

开始展开：

$$
\begin{aligned}
\mathrm{SSE}(\theta)
&= (y - X\theta)^T (y - X\theta) \\
&= y^T y - y^T X\theta - (X\theta)^T y + (X\theta)^T (X\theta)
\end{aligned}
$$

注意到：

$$
y^T X\theta
$$

是一个标量，而标量转置后仍然等于它自己，因此：

$$
y^T X\theta = (X\theta)^T y
$$

所以中间两项可以合并：

$$
\mathrm{SSE}(\theta) = y^T y - 2 y^T X\theta + \theta^T X^T X\theta
$$

现在对 $\theta$ 求梯度：

$$
\nabla_{\theta}(y^T y) = 0
$$

$$
\nabla_{\theta}(-2 y^T X\theta) = -2X^T y
$$

而对于二次型：

$$
\nabla_{\theta}(\theta^T A\theta) = (A + A^T)\theta
$$

这里 $A = X^T X$，而 $X^T X$ 是对称矩阵，所以：

$$
\nabla_{\theta}(\theta^T X^T X\theta) = 2X^T X\theta
$$

因此总梯度为：

$$
\nabla \mathrm{SSE}(\theta) = -2X^T y + 2X^T X\theta
$$

整理得到：

$$
\nabla \mathrm{SSE}(\theta) = 2X^T(X\theta - y)
$$

等价地，也可以写成：

$$
\nabla \mathrm{SSE}(\theta) = -2X^T(y - X\theta)
$$

这正对应你当前代码中的写法。

## 6. 和当前代码变量的对应关系

就是在实现：

$$

$$

只是把数学符号换成了代码变量名。

## 7. 梯度下降更新公式

梯度指向损失上升最快的方向。

因此，如果我们想让损失下降，就应该沿着负梯度方向更新参数：

$$
\theta := \theta - \alpha \nabla \mathrm{SSE}(\theta)
$$

其中：

- $\alpha$ 是学习率
- $\nabla \mathrm{SSE}(\theta)$ 是当前参数位置下的梯度

把梯度代进去：

$$
\theta := \theta - \alpha \cdot 2X^T(X\theta - y)
$$

或者按照你代码里的残差写法：

$$
\theta := \theta - \alpha \cdot \left[-2X^T(y - X\theta)\right]
$$

两者完全等价。

## 8. 如果使用 MSE，会发生什么变化

如果损失函数不是 SSE，而是：

$$
J(\theta) = \frac{1}{2m}\|y - X\theta\|_2^2
$$

那么梯度会变成：

$$
\nabla J(\theta) = \frac{1}{m}X^T(X\theta - y)
$$

你会发现它和 SSE 梯度的区别只在于前面的常数因子。

也就是说：

- `SSE` 和 `MSE` 的最优点相同
- 区别主要体现在梯度大小不同，从而影响学习率怎么调

## 9. 常见回归指标

在线性回归里，我们经常会看到下面几个指标：

- SSE：平方误差和
- MSE：均方误差
- RMSE：均方根误差
- MAE：平均绝对误差

设残差向量为：

$$
r = y - \hat{y}
$$

那么它们的定义分别是：

$$
\mathrm{SSE} = \sum_{i=1}^{m}(y^{(i)} - \hat{y}^{(i)})^2 = \|y - \hat{y}\|_2^2
$$

$$
\mathrm{MSE} = \frac{1}{m}\sum_{i=1}^{m}(y^{(i)} - \hat{y}^{(i)})^2
$$

$$
\mathrm{RMSE} = \sqrt{\mathrm{MSE}}
$$

$$
\mathrm{MAE} = \frac{1}{m}\sum_{i=1}^{m}\left|y^{(i)} - \hat{y}^{(i)}\right|
$$

它们之间的关系可以理解为：

- `SSE`：把所有平方误差直接加起来
- `MSE`：把 `SSE` 再除以样本数 `m`
- `RMSE`：对 `MSE` 再开平方，恢复到和标签相同的量纲
- `MAE`：不用平方，而是直接取绝对值后求平均

## 10. 各指标的优缺点

### 10.1 SSE

优点：

- 形式最直接，和最小二乘推导天然一致
- 写成向量形式非常漂亮：$\mathrm{SSE} = \|y - X\theta\|_2^2$
- 对大误差惩罚更强，适合推导梯度和正规方程

缺点：

- 会随着样本数增大而自然变大，不方便横向比较不同规模数据集
- 单位是“原单位的平方”，解释不够直观

### 10.2 MSE

优点：

- 把误差按样本数平均后，更方便比较不同数据集或不同训练轮次
- 依然保留平方误差的数学性质，适合求导和优化

缺点：

- 单位仍然是“原单位的平方”，解释上不够自然
- 对异常值比较敏感，因为平方会放大大误差

### 10.3 RMSE

优点：

- 和原始标签的量纲一致，最容易直观解释
- 仍然保留了对大误差更敏感的特性

缺点：

- 依然对异常值比较敏感
- 数学推导时不如 SSE / MSE 那么方便

### 10.4 MAE

优点：

- 和原始标签量纲一致
- 对异常值比 MSE / RMSE 更稳一些
- 很适合表达“平均差了多少”这种直觉含义

缺点：

- 数学上不如平方误差那么光滑，求导和优化分析没有那么顺手
- 对大误差的惩罚没有平方误差那么强

### 10.5 怎么选

在你当前这条学习线里，可以先这样记：

- 如果重点是公式推导、正规方程、梯度下降：优先理解 `SSE / MSE`
- 如果重点是结果解释：`RMSE / MAE` 更直观
- 如果担心异常值影响太大：`MAE` 往往更稳
- 如果希望大误差被明显惩罚：`MSE / RMSE` 更合适
