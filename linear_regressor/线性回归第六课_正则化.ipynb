# 线性回归第六课：正则化（Regularization）

这一份 Notebook 专门整理线性回归里的正则化知识，不和梯度推导混在一起。

本节重点：

- 为什么线性回归会引出正则化
- L2 正则化（Ridge 回归）
- L1 正则化（Lasso 回归）
- 为什么 L1 能把参数压到 0，而 L2 通常不会
- 当 $\lambda$ 变大时，模型会发生什么


## 1. 为什么线性回归要引入正则化

在线性回归中，我们希望预测值尽量接近真实值。

如果使用平方误差和（SSE）作为损失函数，那么：

$$
\mathrm{SSE}(\theta) = \sum_{i=1}^{m} \left(y^{(i)} - \hat{y}^{(i)}\right)^2
$$

写成矩阵形式就是：

$$
\mathrm{SSE}(\theta) = \|y - X\theta\|_2^2
$$

但是要注意：训练误差小，不等于模型一定好。

如果模型为了让训练集误差非常小，把某些参数学得特别大，就可能出现过拟合。

所以正则化的核心想法是：

- 不仅要求模型拟合数据
- 还要求模型的参数不要太大

也就是说，我们会在原来的损失函数后面，再加一个参数惩罚项。


## 2. L2 正则化：Ridge 回归

L2 正则化在线性回归中的损失函数通常写成：

$$
J(\theta) = \|y - X\theta\|_2^2 + \lambda \|\theta\|_2^2
$$

展开后可以写成：

$$
J(\theta) = \|y - X\theta\|_2^2 + \lambda \sum_{j=1}^{n} \theta_j^2
$$

这里：

- $\|y - X\theta\|_2^2$ 是原来的误差项
- $\lambda \|\theta\|_2^2$ 是 L2 正则项
- $\lambda$ 控制正则化强度

Ridge 回归的核心作用是：

- 把参数整体压小
- 缓解过拟合
- 但通常不会把参数直接压成 0

如果只看单个参数 $\theta$，惩罚项是：

$$
\lambda \theta^2
$$

它的导数为：

$$
\frac{d}{d\theta}(\lambda \theta^2) = 2\lambda \theta
$$

这说明：

- 当 $|\theta|$ 很大时，惩罚很强
- 当 $|\theta|$ 变小时，惩罚也会变弱
- 当 $\theta$ 接近 0 时，惩罚的“拉回力量”也接近 0

所以 Ridge 更像是：参数越大，压得越狠；参数越小，手越松。


## 3. L1 正则化：Lasso 回归

L1 正则化在线性回归中的损失函数通常写成：

$$
J(\theta) = \|y - X\theta\|_2^2 + \lambda \|\theta\|_1
$$

展开后可以写成：

$$
J(\theta) = \|y - X\theta\|_2^2 + \lambda \sum_{j=1}^{n} |\theta_j|
$$

Lasso 回归的核心作用是：

- 压缩参数
- 缓解过拟合
- 还可能把一部分参数直接压成 0

如果只看单个参数 $\theta$，惩罚项是：

$$
\lambda |\theta|
$$

当 $\theta \neq 0$ 时，它的导数可以写成：

$$
\frac{d}{d\theta}(\lambda |\theta|) =
\begin{cases}
\lambda, & \theta > 0 \\
-\lambda, & \theta < 0
\end{cases}
$$

这说明：

- 只要参数还没有等于 0，它就一直受到一个比较稳定的压缩作用
- 不像 Ridge 那样，越接近 0，惩罚越弱

所以 Lasso 更容易把一些不太重要的参数一路压缩到 0。

因此，L1 正则化会产生稀疏解，并且天然带有一定的特征选择效果。


## 4. 为什么 L1 能压到 0，而 L2 通常不会

这是理解 L1 和 L2 差异最关键的地方。

如果只看单个参数：

- L2 的惩罚项是 $\lambda \theta^2$
- L1 的惩罚项是 $\lambda |\theta|$

它们的导数分别是：

$$
\frac{d}{d\theta}(\lambda \theta^2) = 2\lambda \theta
$$

和：

$$
\frac{d}{d\theta}(\lambda |\theta|) =
\begin{cases}
\lambda, & \theta > 0 \\
-\lambda, & \theta < 0
\end{cases}
$$

从这里可以看出：

### 4.1 L2 的特点

当 $|\theta|$ 变小时，$2\lambda \theta$ 也会变小。

也就是说，参数越接近 0，L2 的压缩作用越弱。

因此它通常只是把参数压得很小，而不是精确地压成 0。

### 4.2 L1 的特点

当 $\theta$ 还没有等于 0 时，L1 在 0 的两侧都会保持一个固定大小的压缩趋势。

也就是说：

- 只要参数还不是 0
- 它就会持续被往 0 的方向推

因此 L1 更容易把参数直接压到 0。

一个很实用的记忆方式是：

- Ridge：越接近 0，手越松
- Lasso：没到 0 之前，不太松手


## 5. 当 $\lambda$ 变大时，Ridge 和 Lasso 会发生什么

$\lambda$ 是正则化强度。

它决定了模型到底有多重视“限制参数大小”这件事。

### 5.1 当 $\lambda = 0$ 时

如果：

$$
\lambda = 0
$$

那么无论是 Ridge 还是 Lasso，都会退化成普通线性回归。

### 5.2 当 $\lambda$ 逐渐变大时，Ridge 的变化

- 所有参数会被逐渐压小
- 模型会越来越保守
- 训练误差通常会上升
- 如果 $\lambda$ 太大，模型会开始欠拟合

### 5.3 当 $\lambda$ 逐渐变大时，Lasso 的变化

- 参数会被压缩
- 会有越来越多的参数变成 0
- 模型会越来越稀疏
- 如果 $\lambda$ 太大，同样会欠拟合

### 5.4 从训练误差和测试误差看

通常随着 $\lambda$ 增大，会出现这样的趋势：

- 训练误差上升
- 测试误差可能先下降，再上升

原因是：

- 一开始，适当正则化可以缓解过拟合
- 但正则化太强，又会导致欠拟合

所以结论不是“$\lambda$ 越大越好”，而是：

$$
\lambda \text{ 要取一个合适的值}
$$


## 6. 本节小结

现在可以把正则化在线性回归中的作用总结成下面几句话：

1. 正则化是在原损失函数后面再加一个参数惩罚项。
2. L2 正则化对应 Ridge 回归，主要作用是让参数整体变小。
3. L1 正则化对应 Lasso 回归，除了压缩参数，还可能把一部分参数直接压成 0。
4. Ridge 更偏向“收缩参数”，Lasso 更偏向“收缩 + 特征选择”。
5. $\lambda$ 控制正则化强度，过小可能抑制不住过拟合，过大又可能导致欠拟合。
